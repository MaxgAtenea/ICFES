```python
import pandas as pd
```

Este notebook es para pegar los resultado del valor agregado, abriendo por programa, con la informacion de snies. el resultado se le envió a Julian por whatsapp el 16 de julio. Esto fue para la presentación de Victor del 17 de julio.

```python
snies_file = "30062025 SNIES Programas.xlsx"
resultados_file = "va_cine_especifico_2022_2023_run15072025.xlsx"

snies_path = "../../data/SNIES_CINE_raw/"
resultados_path = "../../data/Resultados_VA/cine_especifico/bogota_region/"
```

```python
# Cargar los archivos
snies = pd.read_excel(snies_path + snies_file)
resultados = pd.read_excel(resultados_path + resultados_file)

# Expandir los códigos SNIES separados por ";"
resultados_expandidos = (
    resultados
    .assign(CÓDIGO_SNIES_DEL_PROGRAMA=resultados['snies_programas']
            .astype(str)
            .str.split(';'))
    .explode('CÓDIGO_SNIES_DEL_PROGRAMA')
)

# Limpiar espacios y convertir a numérico si corresponde
resultados_expandidos['CÓDIGO_SNIES_DEL_PROGRAMA'] = (
    resultados_expandidos['CÓDIGO_SNIES_DEL_PROGRAMA']
    .str.strip()
)

if pd.api.types.is_numeric_dtype(snies['CÓDIGO_SNIES_DEL_PROGRAMA']):
    resultados_expandidos['CÓDIGO_SNIES_DEL_PROGRAMA'] = pd.to_numeric(
        resultados_expandidos['CÓDIGO_SNIES_DEL_PROGRAMA'], errors='coerce'
    )

# Realizar el left join
resultado_final = resultados_expandidos.merge(
    snies,
    on=['CÓDIGO_SNIES_DEL_PROGRAMA'],
    how='left'
)

# Ver resultado
resultado_final.head()
```

```python
cols_export = [
    'icine',
    'nombre_institucion',
    'codigo_institucion',
    'coeficiente_PG',
    'coeficiente_LC',
    'coeficiente_RC',
    'periodo',
    'CÓDIGO_SNIES_DEL_PROGRAMA',
    'NOMBRE_DEL_PROGRAMA',
    'NÚCLEO_BÁSICO_DEL_CONOCIMIENTO',
    'MUNICIPIO_OFERTA_PROGRAMA',
]
```

```python
resultado_final[cols_export].reset_index(drop=True).to_excel("valor_agregado_xprograma_2022_2023.xlsx",index=False)
```
