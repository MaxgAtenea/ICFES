```python
import pandas as pd

# 1. Cargar archivos
res_2021_2022 = pd.read_excel("valor_agregado_xprograma_2021_2022.xlsx")
res_2022_2023 = pd.read_excel("valor_agregado_xprograma_2022_2023.xlsx")

# 2. Eliminar columna 'periodo' si existe
res_2021_2022 = res_2021_2022.drop(columns=['periodo',"INBC"], errors='ignore')
res_2022_2023 = res_2022_2023.drop(columns=['periodo',"INBC","NÚCLEO_BÁSICO_DEL_CONOCIMIENTO"], errors='ignore')

# 3. Estandarizar nombres de columnas
```

```python
# Para res_2021_2022
res_2021_2022 = res_2021_2022.rename(columns={
    'nombre_del_programa': 'nombre_del_programa',
    'municipio_oferta_programa': 'municipio_oferta_programa',
    'va_razona_cuantitat': 'coeficiente_rc',
    'va_lectura_critica': 'coeficiente_lc',    
    'icine_especifico': 'icine_especifico',
    'codigo_institucion': 'codigo_institucion',
    'nombre_institucion': 'nombre_institucion',
    "codigo_snies_del_programa": "codigo_snies"
})

# Para res_2022_2023
res_2022_2023 = res_2022_2023.rename(columns={
    'NOMBRE_DEL_PROGRAMA': 'nombre_del_programa',
    'MUNICIPIO_OFERTA_PROGRAMA': 'municipio_oferta_programa',
    'coeficiente_RC': 'coeficiente_rc',    
    'coeficiente_LC': 'coeficiente_lc',
    'coeficiente_PG': "coeficiente_pg",
    'icine': 'icine_especifico',    
    'codigo_institucion': 'codigo_institucion',
    'nombre_institucion': 'nombre_institucion',
    'CÓDIGO_SNIES_DEL_PROGRAMA': "codigo_snies"
})
```

```python
# 4. Realizar el outer join sobre la llave estandarizada 'codigo_snies'
resultado = pd.merge(
    res_2021_2022,
    res_2022_2023,
    on=['codigo_snies',"nombre_institucion","codigo_institucion","icine_especifico"],
    how='outer',
    suffixes=('_2021_2022', '_2022_2023')
)
```

```python
resultado.to_excel("resultados_va_icfes_y_propios.xlsx",index=False)
```

```python
columnas_prioritarias = [
    "codigo_institucion",
    "nombre_institucion",
    "codigo_snies",
    "icine_especifico",
    "nombre_del_programa_2021_2022",
    "nombre_del_programa_2022_2023",
    "municipio_oferta_programa_2021_2022",
    "municipio_oferta_programa_2022_2023",
]
otras_columnas = [col for col in resultado.columns if col not in columnas_prioritarias]
nuevo_orden = columnas_prioritarias + otras_columnas

resultado = resultado[nuevo_orden]

resultado.to_excel("resultados_va_icfes_y_propios.xlsx", index=False)
```
